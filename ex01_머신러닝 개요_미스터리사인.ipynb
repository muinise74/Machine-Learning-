# 미스터리 사인
- 주어진 수를 보고 규칙을 찾는 문제

## Question 1
- 큰 수 + 두 수의 차
- 22 ? 7 = 37

### 데이터 만들기

```python
list1 = []
for i in range(1,200):
    for j in range(1,200):
        result = (i+(i-j)) if i > j else (j+(j-i))
        list1.append([i,j,result])
```

```python
for i in list1:
    print(i,end=", ")
```

### 데이터 전처리

```python
import pandas as pd
data1 = pd.DataFrame(list1, columns=['N1','N2','Result'])
data1.head()
```

```python
data1
```

### 문제와 정답 분리
- X: 문제
- Y: 정답

```python
X = data1.loc[:,['N1','N2']]
Y = data1.loc[:,'Result']
```

```python
X
```

```python
Y
```

### 머신 러닝 모델 불러오기

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

### 모델 사용하기(학습)
- 규칙 찾기

```python
rf.fit(X,Y)
```

### 모델 사용하기(예측)

```python
rf.predict([[22,7]])
```

```python
rf.predict([[230,86]])
```

## Question 2
- 큰 수 % 작은 수
- 4 ? 19 = 3

### 데이터 만들기

```python
list2 = []
for i in range(1,1001):
    for j in range(1,1001):
        result = i%j if i > j else j%i
        list2.append([i,j,result])
```

### 데이터 전처리

```python
data2 = pd.DataFrame(list2, columns=['N1','N2','Result'])
data2
```

### 문제와 정답 분리

```python
X2 = data2.loc[:,['N1','N2']]
Y2 = data2.loc[:,'Result']
```

```python
X2
```

```python
Y2
```

### 머신 러닝 모델 불러오기

```python
rf2 = RandomForestRegressor()
```

### 모델 사용하기(학습)

```python
rf2.fit(X2,Y2)
```

### 모델 사용하기(예측)

```python
rf2.predict([[22,7]])
```

```python
rf2.predict([[2234,234]])
```

```python
2234%234
```

```python
rf2.predict([[1234,234]])
```
